```python
# 보간이 잘 되고 있는지 확인하기 위해 랜덤으로 5개의 샘플 추출
random_rows = data.iloc[:, 274:].sample(n=5, random_state=1)
random_rows.head()
```

```python
random_index_list = [data.index.get_loc(idx) for idx in random_rows.index]

print(random_index_list)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(24,10))
for idx in random_index_list:
  plt.plot(data.iloc[idx, 274:].index, data.iloc[idx, 274:].values, label=f'Row {idx}')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Randomly Selected Rows')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

# 머신러닝 적용
- 감귤의 착과량 예측
- RandomForestRegressor만 사용


```python
train_x, test_x, train_y, test_y = train_test_split(data.iloc[:,1:] , data['착과량(int)'], test_size=0.3, random_state=1740)
```

```python
print(train_x.shape)
print(test_x.shape)

print(train_y.shape)
print(test_y.shape)
```

```python
randomf_reg = RandomForestRegressor(random_state=1740,
                                    max_depth=5,
                                    n_estimators=200,
                                    max_features = 0.5)
randomf_reg.fit(train_x, train_y)
```

```python
from sklearn.metrics import mean_squared_error

# 예측 생성
predictions = randomf_reg.predict(test_x)

# 성능 평가
mse = mean_squared_error(test_y, predictions)

# 결과 출력
print(f"Mean Squared Error: {mse}")
```
